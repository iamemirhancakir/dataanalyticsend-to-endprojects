```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
```

# En Çok Satılan Ürünler - Matplotlib - Bar chart

```python
top_products = {
    "Urun": [
        "Çamaşır Makinesi", "Tansiyon Aleti", "Yazıcı", "Akıllı Ampul",
        "Elektrikli Izgara", "Mikser", "Tablet", "Blender",
        "Termostat", "Anakart"
    ],
    "Satis_Tutari": [71367578, 71209703, 71171929, 71042559, 70879561, 70844239, 70755072, 70614095, 70236235, 69422363]
}

products_df = pd.DataFrame(top_products)

plt.figure(figsize=(10, 6))
plt.barh(products_df['Urun'], products_df['Satis_Tutari'], color="skyblue")
plt.title("En Çok Satılan Ürünler", fontsize=16)
plt.xlabel("Toplam Satış", fontsize=12)
plt.ylabel("Ürünler", fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/top_products.png", dpi=300)
plt.show()
```

# Aylık Satış Analizi - Seaborn - Line plot ve Plotly - Interactive line plot

```python
monthly_sales = {
    "Ay": [
        "Ocak", "Şubat", "Mart", "Nisan", "Mayıs", "Haziran",
        "Temmuz", "Ağustos", "Eylül", "Ekim", "Kasım", "Aralık"
    ],
    "Satis_Tutari": [
        292317291, 264892462, 290319543, 277679394, 289302895,
        278079951, 292552989, 284865450, 285671914, 287835101,
        277563446, 287898781
    ]
}

monthly_sales_df = pd.DataFrame(monthly_sales)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=monthly_sales_df,
    x="Ay", y="Satis_Tutari",
    marker="o", color="blue"
)
plt.title("Aylık Satış Analizi", fontsize=16)
plt.xlabel("Ay", fontsize=12)
plt.ylabel("Toplam Satış", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/monthly_sales.png", dpi=300)
plt.show()
```

```python
fig = px.line(
    monthly_sales_df, x="Ay", y="Satis_Tutari",
    title="Aylık Satış Analizi",
    markers=True
)
plt.savefig("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/monthly_sales_plotly.png", dpi=300)
fig.show()
```

# Kategori Bazlı En Çok Satan Ürünler - Matplotlib - Pie chart ve Plotly - Interactive pie chart

```python
category_sales = {
    "Kategori": [
        "Elektronik", "Küçük Ev Aletleri", "Bilgisayar Parçaları",
        "Beyaz Eşya", "Sağlık", "Ev Aksesuarları"
    ],
    "Satis_Tutari": [
        1429927974, 957669394, 478225937, 339196192, 137890166, 66069554
    ]
}

category_df = pd.DataFrame(category_sales)

plt.figure(figsize=(8, 8))
plt.pie(
    category_df["Satis_Tutari"],
    labels=category_df["Kategori"],
    autopct="%1.1f%%",
    colors=sns.color_palette("pastel"),
    startangle=140
)
plt.title("Kategori Bazlı En Çok Satan Ürünler", fontsize=16)
plt.savefig("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/category_sales.png", dpi=300)
plt.show()
```

```python
fig = px.pie(
    category_df,
    values="Satis_Tutari",
    names="Kategori",
    title="Kategori Bazlı En Çok Satan Ürünler",
    color_discrete_sequence=px.colors.sequential.RdBu
)
fig.show()
```

# Aylara Göre Müşteri Sayısı - Seaborn - Bar plot ve Plotly - Interactive bar chart

```python
monthly_customers = {
    "Ay": [
        "Mart", "Ağustos", "Mayıs", "Temmuz", "Ekim", "Ocak", "Aralık",
        "Eylül", "Nisan", "Kasım", "Haziran", "Şubat"
    ],
    "Musteri_Sayisi": [6703, 6690, 6686, 6684, 6684, 6648, 6643, 6567, 6557, 6519, 6504, 6385]
}

customers_df = pd.DataFrame(monthly_customers)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=customers_df,
    x="Ay", y="Musteri_Sayisi",
    palette="viridis"
)
plt.title("Aylara Göre Müşteri Sayısı", fontsize=16)
plt.xlabel("Ay", fontsize=12)
plt.ylabel("Müşteri Sayısı", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/monthly_customers.png", dpi=300)
plt.show()
```

```python
fig = px.bar(
    customers_df,
    x="Ay", y="Musteri_Sayisi",
    title="Aylara Göre Müşteri Sayısı",
    color="Musteri_Sayisi",
    color_continuous_scale="viridis"
)
fig.show()
fig.write_image("C:/Users/lenovo/OneDrive/Desktop/dataanalyticsproject/Visuals/monthly_customers_plotly.png", engine='kaleido', format="png", width=800, height=600)
```
